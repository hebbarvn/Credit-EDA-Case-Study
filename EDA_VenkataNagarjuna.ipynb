# Binning the Columns

```python
#Binning the AMT_CREDIT 
#According to the Quartile and removing the outlier
l1= list(appl_data['AMT_CREDIT'].quantile([0.25,0.5,0.75,0.99]).astype(int))
def Cat(x):
    if (x<=l1[0]):
        return 'Low'
    elif (x>l1[0]) & (x<=l1[1]):
        return 'Medium'
    elif (x>l1[1]) & (x<=l1[2]):
        return 'High'
    elif (x>l1[2]) & (x<=l1[3]):
        return 'Very_High'
    else:
        return 'Outlier'
appl_data['Crd_Cat_Req'] = appl_data['AMT_CREDIT'].apply(lambda x: Cat(x))
outlier_crd=appl_data[appl_data['Crd_Cat_Req']=='Outlier']
outlier_crd
```

```python
#Plotting Amt Credit after removing the outliers
plt.suptitle("Amount Credit beforetreating Outliers", size=16)
sns.countplot(x='Crd_Cat_Req',data=appl_data)
```

```python
#Binning the AMT_ANUITY According to the Quartile and finding the outlier 
l1= list(appl_data['AMT_ANNUITY'].quantile([0.25,0.5,0.75,0.99]).astype(int))

appl_data['Cat_ANNUITY'] = appl_data['AMT_ANNUITY'].apply(lambda x: Cat(x))
Outlier_Annu=appl_data[appl_data['Cat_ANNUITY']=='Outlier']
Outlier_Annu
```

```python
#Plotting Amount Annuity 
plt.suptitle("Amount Annuity before treating Outliers", size=16)
sns.countplot(x='Cat_ANNUITY',data=appl_data)
```

```python
#Binning the AMT_INCOME_TOTAL 
#According to the Quartile and finding the outlier 
l1= list(appl_data['AMT_INCOME_TOTAL'].quantile([0.25,0.5,0.75,0.995]).astype(int))
print(l1)
```
